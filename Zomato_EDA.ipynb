## ZOMATO DATASET EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('zomato.csv', encoding='latin-1')
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum() > 1]
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

```python
df_final=pd.merge(df,df_country, on='Country Code', how='left')
df_final.head(2)
```

```python
country_names=df_final.Country.value_counts().index
```

```python
country_values=df_final.Country.value_counts().values
```

```python
# Country names and values on Pie Chart, show which country do bigger transactions.
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

OBSERVATION: Zomato maximum transactions from India, then USA and 3rd UK

```python
rating = df_final.groupby(['Aggregate rating','Rating color','Rating text','Country']).size().reset_index().rename(columns={0:'Ratings Count'})
```

```python
rating
```

OBSERVATION:
    
    1. When Rating is between 4.5 to 4.9---> Excellent
    2. When Rating are between 4.0 to 3.4--->very good
    3. when Rating is between 3.5 to 3.9----> good
    4. when Rating is between 3.0 to 3.4----> average
    5. when Rating is between 2.5 to 2.9----> average
    6. when Rating is between 2.0 to 2.4----> Poor

```python
# Maximum Ratings given by Country
rating.groupby('Ratings Count')['Country'].max()
```

```python
rating.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,6)
sns.barplot(x="Aggregate rating", y="Ratings Count", data = rating)
```

```python
sns.barplot(x="Aggregate rating", y="Ratings Count", hue="Rating color", data = rating, palette=['white','red','orange','yellow','green','green'])
```

## OBSERVATION:
1. Not rated is very high
2. Maximum Rating between 2.5 to 3.4

```python
sns.countplot(x='Rating color', data=rating, palette=['blue','red','orange','yellow','green','green'])
```

```python
## Countries which gave 0 ratings
df_final[df_final['Rating color']=='White'].groupby('Country').size().reset_index()
```

```python
# Find out which country have which currency?
df_final.groupby('Country')['Currency'].max()
```

```python
df_final[['Country', 'Currency']].groupby(['Country', 'Currency']).size().reset_index().rename(columns={0:'Currency Values'})
```
